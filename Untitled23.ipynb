```python
import pandas as pd
import numpy as np 
```

```python
ds=pd.Series([10,20,30,40],index=['a','b','c','d'])
print(ds)
```

```python
ds=pd.Series({'v':10,'h':20,'a':30})
print(ds)
```

```python
ds=pd.Series((10,20,30,40))
print(ds)
```

```python
ds=pd.Series(range(10))
print(ds)
```

```python
ds=pd.Series(np.array([1,2,3,4,5]))
print(ds)
```

```python
ds=pd.DataFrame([['v','h','a'],['p','s','s'],['p','t','s']],index=['a','b','c'],columns=['p','q','r'])
print(ds)
```

```python
ds=pd.DataFrame({'p':['v','h','a'],'q':['p','s','s'],'r':['p','t','s']},index=['a','b','c'])
print(ds)
```

```python
ds=pd.DataFrame(np.array(['A','B','c','d']),index=[1,2,3,4])
print(ds)
```
