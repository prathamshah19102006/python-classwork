```python
import pandas as pd 
import numpy as np
import plotly.express as px
import plotly.io as pio
pio.renderers.default='notebook'
```

```python
df=pd.read_csv('social_media_engagement.csv')
```

```python
px.bar(x=df.post_type,y=df.likes,width=1000,height=1000,title='post type vs likes',text_auto=True,color=df.platform,color_discrete_map={'Facebook':'black'
                                                                                                                                       ,'Instagram':'red','Twitter':'blue'})
```

```python
px.bar(x=df.post_type,y=df.likes,width=1000,height=1000,title='post type vs likes',text_auto=True,color=df.comments,color_discrete_map={'Facebook':'black'
                                                                                                                                       ,'Instagram':'red','Twitter':'blue'})
```

```python
px.bar(x=df.post_type,y=df.likes,width=1000,height=1000,title='post type vs likes',text_auto=True,color=df.comments,hover_data={'A':df.post_day,'B':df.sentiment_score})
```

```python
px.bar(x=df.likes,y=df.post_type,width=1000,height=1000,title='post type vs likes',text_auto=True,color=df.comments,color_discrete_map={'Facebook':'black'
                                                                                                                                       ,'Instagram':'red','Twitter':'blue'},
      orientation='h')
```

```python
px.bar(x=df.likes,y=df.post_type,width=1000,height=1000,title='post type vs likes',text=df.likes,color=df.comments,color_discrete_map={'Facebook':'black'
                                                                                                                                       ,'Instagram':'red','Twitter':'blue'},
      orientation='h')
```

```python
px.bar(y=df.likes,x=df.post_type,width=1000,height=1000,title='post type vs likes',hover_data={'A':df.post_day,'B':df.sentiment_score},text=df.likes,color=df.platform,pattern_shape=df.sentiment_score)
```

```python
px.bar(y=[df.likes,df.shares,df.comments],x=df.post_type,barmode='stack')
```

```python
px.bar(y=[df.likes,df.shares,df.comments],x=df.post_type,barmode='group')
```

```python
df['comments_n']=df['comments']*-1
```

```python
px.bar(x=df.post_id,y=[df.comments_n,df.shares],barmode='relative')
```

```python
px.scatter(x=df.comments,y=df.likes,title='VHA',height=1000,width=1000,color=df.shares,symbol=df.platform)
```

```python
px.scatter(x=df.comments,y=df.likes,title='VHA',height=1000,width=1000,size=df.shares,color=df.platform)
```

```python
px.scatter(x=df.comments,y=df.likes,title='VHA',height=1000,width=1000,size=df.shares,color=df.post_type,symbol=df.post_type)
```

```python
px.scatter(x=df.comments,y=df.likes,title='VHA',height=1000,width=1000,size=df.shares,color=df.post_type,facet_row=df.post_day,
          facet_col=df.sentiment_score)
```

```python
px.scatter_3d(x=df.likes,y=df.comments,z=df.shares,color=df.platform,size=df.likes)
```
