```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
df=pd.read_csv('social_media_engagement.csv')
df
```

```python
def decide(x):
    if x>3000:
        return 'good'
    return 'bad'
```

```python
df['likes'].apply(decide)
```

```python
def danger(x):
    if x['likes']>500 and x['comments']<100:
        return 'risk high'
    elif x['likes']>2000:
        return 'medium'
    elif x['shares']>1000:
        return 'low'
    else:
        return 'not eligible'
    
```

```python
df['risk']=df.apply(danger,axis=1)
df
```
