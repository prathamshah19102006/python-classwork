```python
import numpy as np
import pandas as pd
```

```python
df=pd.DataFrame(np.random.randint(1,100,(20,5)),columns=['A','B','C','D','E'])
df
```

```python
df.apply(['sum','mean'])
```

```python
df.apply(sum,axis=1)
```

```python
df.apply(lambda x:x['A']+x['B']+x['C'],axis=1)
```

```python
df['total']=df.apply(lambda x:x['A']+x['B']+x['C'],axis=1)
df
```

```python
def fun(df):
    for i in df['total']:
        if i>300:
            print('A+')
        elif(i>250):
            print('A')
        elif(i>200):
            print('B')
        else:
            print('F')
fun(df)
```

```python
def grade(x):
    if x>300:
        return 'A+'
    if x>250:
        return 'A'
    if x>200:
        return 'B'
    return 'F'
```

```python
df["grade"]=df['total'].apply(grade)
df
```

```python
def result(x):
    for i in x:
        if x<35:
            return 'Fail'
        else:
            return 'pass'
```

```python
df['result']=df.apply(result,axis=1)
```

```python
def res(x):
    if x['A']<35 or x['B']<35 or x['C']<35 or x['D']<35 or x['E']<35:
        return 'fail'
    return 'pass'
```

```python
df['result']=df.apply(res,axis=1)
df
```
